```python
features_with_nan=[features for features in dataset.columns if dataset[features].isnull().sum() > 1]
```

```python
features_with_nan
```

```python
for feature in features_with_nan:
    print(feature,np.round(dataset[feature].isnull().mean(),5)*100, '%missing values')
```

```python
## lets plot some amazing diagrams ##
for feature in features_with_nan:
    data=dataset.copy()
    ## lets make  a variable that indicate 1 if the observation is missing 
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```
