```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/shreyas219/breast_cancer_classification/master/breast_cancer.csv')
```

```python
df
```

```python
df.head()
```

```python
df['target'] = df['target'].astype(int)
```

```python
sns.countplot(df['target'])
```

```python
sns.pairplot(df,hue ='target',vars=['mean radius','mean texture','mean area','mean perimeter','mean smoothness'])
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(),annot=True)
```

```python
y = df['target']
X = df.drop('target', axis = 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.svm import SVC
```

```python

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train,y_train)
```

```python
y_pred = svc_model.predict(X_test)
```

```python
model_filename = 'model.pkl'
joblib.dump(svc_model,model_filename)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```
