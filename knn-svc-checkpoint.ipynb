```python
import pandas as pd 
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree 
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.metrics import accuracy_score , confusion_matrix , roc_auc_score , roc_curve
from pandas_profiling import ProfileReport
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('winequality-red (1).csv')
```

```python
df
```

```python
x = df.drop(columns=['quality'])
```

```python
y = df.quality
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=30)
```

```python
bag_dt = BaggingClassifier(DecisionTreeClassifier(),n_estimators=10)
```

```python
bag_dt.fit(x_train, y_train)
```

```python
bag_dt.predict(x_test)
```

```python
bag_dt.base_estimator_
```

```python
bag_dt.classes_
```

```python
bag_dt.estimator_params
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
bag_knn = BaggingClassifier(KNeighborsClassifier(6),n_estimators=10)
```

```python
bag_knn.fit(x_train,y_train)
```

```python
bag_knn.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

```python
bag_dt.score(x_test,y_test)
```

```python
bag_knn.score(x_test,y_test)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
pram  = {
    'n_neighbors':[3,5,7,9,12,13,15,17,21],
    'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size' : [10 , 15 , 20 , 25 , 30 , 35 , 45 , 50 ],
    'p' : [1,2],
    'weights' : ['uniform', 'distance']
    
    
    
}
```

```python
grid_cv = GridSearchCV(knn,param_grid=pram)
```

```python
grid_cv.fit(x_train,y_train)
```

```python
grid_cv.best_params_
```
